7- What are the advantages of using generators over regular functions?
- Generators are memory-efficient because they produce values one at a time instead of storing the whole result in memory. They make code faster for large datasets and support infinite sequences easily. Generators also use simple 'yield' syntax, making them cleaner than writing custom iterator classes. Additionally, they improve performance by generating values only when needed (lazy evaluation).

9- Explain the purpose and usage of the 'map()' function in Python.
- 'map()' in Python is used to apply a function to every item in an iterable (like a list or tuple). It returns a 'map' object, which is an iterator containing the results. You usually convert it to a list using 'list(map(...))' to see the output clearly.
- 'map()' helps avoid writing loops again and again, making code cleaner. It is especially useful when combined with functions like 'lambda' for quick operations.

10- What is the difference between 'map()', 'reduce()', and 'filter()' functions in Python?
- **'map()' - Transform each element**

  * Takes a function and an iterable (list, tuple, etc.).
  * Applies the function to *every element* and returns a new iterator of transformed values.
  * Used when you want to change or modify all items.<br>
  Example: converting all names to uppercase, squaring all numbers.

- **'filter()' - Keep only the elements that meet a condition**

  * Takes a function that returns True/False and an iterable.
  * Only the elements for which the function returns True are kept.
  * Used when you want to select or filter out items.<br>
  Example: keeping only numbers > 10, filtering odd numbers.

- **'reduce()' - Combine elements into a single value**

  * Comes from 'functools' ('from functools import reduce').
  * Applies a function continuously to accumulate values into one final result.
  * Used when you want to summarize or reduce a list.<br>
  Example: sum of all numbers, product of all values, finding maximum.

```python
# 1- Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even(numbers):
  sum = 0
  for i in numbers:
    if i%2 == 0:
      sum += i
  return sum

sum_of_even([1,2,3,4,5,6])
```

```python
# 2- Create a Python function that accepts a string and returns the reverse of that string.
def reverse_str(text):
  return text[::-1]

reverse_str('hello')
```

```python
# 3- Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
  lis = []
  for i in numbers:
    lis.append(i**2)
  return lis

square_numbers([1,2,3,4,5])
```

```python
# 4- Write a Python function that checks if a given number is prime or not from 1 to 200.
number = int(input("enter a number between 1 to 200: "))
def is_prime(number):
  if number < 2:
    return 'not prime'
  for i in range(2, number):
    if number % i == 0:
      return 'not prime'
  return 'prime'

is_prime(30)
```

```python
# 5- Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class print_fib():
  def __init__(self, n):
    self.n = n
    self.count = 0
    self.a = 0
    self.b = 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.count >= self.n:
      raise StopIteration

    if self.count == 0:
      self.count += 1
      return 0

    elif self.count == 1:
      self.count += 1
      return 1

    val = self.a + self.b
    self.a, self.b = self.b, val
    self.count += 1
    return val

fib = print_fib(10)
for i in fib:
  print(i)
```

```python
# 6- Write a generator function in Python that yields the powers of 2 up to a given exponent.
def generator_func(number):
  for i in range(1,number+1):
    yield i**2

square = generator_func(10)
for i in square:
  print(i)
```

```python
# 7- Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(file):
  with open(file, 'r') as f:
    for line in f:
      yield line

file_read = read_file('text.txt')
for i in file_read:
  print(i)
```

```python
# 8- Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
lis = [(1,2), (3,4), (5,6), (7,8)]
sorted_list = sorted(lis, key=lambda x: x[1])
print(sorted_list)
```

```python
# 9- Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [1,2,3,4,5]
fahrenheit = list(map(lambda x: x*9/5 +32, celsius))
print(fahrenheit)
```

```python
# 10- Create a Python program that uses `filter()` to remove all the vowels from a given string.
vowels = "aeiouAEIOU"
text = 'hello world'
lis = ""
a = filter(lambda x: x not in vowels, text)
for i in a:
  lis += i
print(lis)
```
